# Bipartite Graphs

![](images/bipartite1.png)

Bipartite Graph is a graph whose nodes can be split into two sens L and R and every edge conncets and node in L with a node in R.

```python
import networkx as nx
from networkx.algorithms import bipartite
%matplotlib inline


B = nx.Graph()
B.add_nodes_from(['A','B','C','D','E'], bipartite=0) # label one set of nodes 0
B.add_nodes_from([1,2,3,4], bipartite=1)
B.add_edges_from([('A',1),('B',1),('C',1),('C',3),('D',2),('E',3),('E',4)])

nx.draw_networkx(B)
```

## Check if a graph is bipartite - 1

```python
bipartite.is_bipartite(B)
```

```python
B.add_edge('A','B')
```

```python
bipartite.is_bipartite(B)
```

```python
B.remove_edge('A','B')
```

## Check if a graph is bipartite -2

```python
X=set([1,2,3,4])
bipartite.is_bipartite_node_set(B,X)
```

```python
X=set(['A','B','C','D','E'])
bipartite.is_bipartite_node_set(B,X)
```

```python
X=set([1,2,3,4,'A'])
bipartite.is_bipartite_node_set(B,X)
```

## Getting each set of nodes of a bipartite graph:

```python
bipartite.sets(B)
```

# Projected Graphs

![](images/bipartite2.png)

**L-Bipartite Graph Projection:** Network of nodes in group L, where a pair of nodes is connceted if they have a common neighbor in R in the bipartite graph. Similar definition for R-Bipartite graph projection

![](images/bipartite3.png)

```python
B=nx.Graph()
B.add_edges_from([('A',1),('B',1),('C',1),('D',1),('H',1),('B',2),('C',2),('D',2),('E',2),('G',2),('E',3),('F',3),('H',3),('J',3),('E',4),('I',4),('J',4)])
X=set(['A','B','C','D','E','F','G','H','I','J'])
P1=bipartite.projected_graph(B,X)

nx.draw_networkx(P1)
```
